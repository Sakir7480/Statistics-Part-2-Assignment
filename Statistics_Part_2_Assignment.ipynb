1.What is hypothesis testing in statistics ?

.Hypothesis testing is a statistical method used to make decisions or inferences about a population based on a sample of data. It involves formulating two competing hypotheses and using sample data to determine which one is more likely to be true.

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

.The null hypothesis (Hâ‚€) and the alternative hypothesis (Hâ‚ or Ha) are fundamental concepts in hypothesis testing.

.Null Hypothesis (Hâ‚€)

The null hypothesis represents the default assumption or the status quo.

It states that there is no effect, no difference, or no relationship between variables.

Researchers assume Hâ‚€ to be true unless there is strong evidence against it.

It is usually tested to be rejected or not rejected (never "accepted").

3.What is the significance level in hypothesis testing, and why is it important?

.Significance Level (Î±) in Hypothesis Testing

The significance level (Î±) is the probability of rejecting the null hypothesis (Hâ‚€) when it is actually true. In other words, it represents the risk of making a Type I error (a false positive).

.Why is Significance Level Important?

.Controls the Probability of a Type I Error

A Type I error occurs when we incorrectly reject a true Hâ‚€.

Example: Concluding that a new drug is effective when it actually isnâ€™t.

.Balances Sensitivity and Specificity

A lower Î± reduces the chance of a Type I error but increases the chance of a Type II error (failing to reject a false Hâ‚€).

A higher Î± increases sensitivity but raises the risk of false positives.

4.What does a P-value represent in hypothesis testing?

.A p-value is the probability of obtaining a test statistic as extreme as (or more extreme than) the observed results, assuming that the null hypothesis (Hâ‚€) is true.

.Interpreting the P-Value

A small p-value (typically â‰¤ significance level Î±) suggests that the observed data is unlikely under Hâ‚€, leading to rejection of Hâ‚€.

A large p-value (typically > Î±) suggests that the observed data is consistent with Hâ‚€, meaning we fail to reject it.

5.How do you interpret the P-value in hypothesis testing?

.Interpreting the P-Value in Hypothesis Testing

The p-value helps determine whether the observed data provides enough evidence to reject the null hypothesis (Hâ‚€).

.General Interpretation Rules:

Small P-Value (p â‰¤ Î±): Reject Hâ‚€

The data is unlikely under Hâ‚€.

Strong evidence suggests that Hâ‚€ is false, so we support the alternative hypothesis (Hâ‚).

Example: If p = 0.02 and Î± = 0.05, we reject Hâ‚€.

6.What are Type 1 and Type 2 errors in hypothesis testing?

.Type I and Type II Errors in Hypothesis Testing

In hypothesis testing, two types of errors can occur when making a decision about the null hypothesis (Hâ‚€).

.Type I Error (False Positive)

Occurs when Hâ‚€ is true, but we incorrectly reject it.

It is equivalent to a false alarmâ€”detecting an effect when there is none.

The probability of making a Type I error is represented by the significance level (Î±).

.Type II Error (False Negative)

Occurs when Hâ‚€ is false, but we fail to reject it.

It is equivalent to missing a real effectâ€”failing to detect a difference when one exists.

The probability of making a Type II error is represented by Î² (beta).

7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

.One-Tailed vs. Two-Tailed Tests in Hypothesis Testing

A one-tailed test and a two-tailed test refer to the direction of the hypothesis test and how we assess statistical significance.

.One-Tailed Test (Directional Test)

Tests whether the effect is only in one direction (either greater than or less than).

The alternative hypothesis (Hâ‚) specifies a direction (e.g., greater than or less than).

The entire significance level (Î±) is concentrated in one tail of the probability distribution.

.Two-Tailed Test (Non-Directional Test)

Tests whether the effect is in either direction (greater than or less than).

The alternative hypothesis (Hâ‚) states that there is a difference, but does not specify direction.

The significance level (Î±) is split between both tails of the probability distribution

10.What is the T-distribution, and when should it be used instead of the normal distribution?

.The T-distribution (also called Studentâ€™s T-distribution) is a probability distribution that is similar to the normal distribution but has heavier tails. This makes it more appropriate for smaller sample sizes. As the sample size increases, the T-distribution becomes closer to the normal distribution.

Key Characteristics of the T-Distribution

11.What is the difference between a Z-test and a T-test?

.The Z-test and T-test are both statistical tests used in hypothesis testing to assess whether there is a significant difference between a sample and a population (or between two samples), but they are used under different conditions.

.Population Standard Deviation (Ïƒ)

Z-Test: Used when the population standard deviation (Ïƒ) is known.

T-Test: Used when the population standard deviation is unknown, and the sample standard deviation (s) is used instead.

Sample Size

Z-Test: Typically used when the sample size is large (n â‰¥ 30). The Central Limit Theorem ensures that the sampling distribution of the sample mean approaches normality, regardless of the underlying population distribution, as the sample size increases.

T-Test: Used when the sample size is small (n < 30). For small samples, the distribution of the sample mean is more variable, and the T-distribution accounts for this increased uncertainty.

14.What is a confidence interval, and how is it used to interpret statistical results?

.A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate an unknown population parameter (such as the population mean, proportion, or difference between two groups). The confidence interval provides an estimated range of where the true population parameter is likely to fall, along with a degree of confidence that this interval contains the true value.

.Interpreting a Confidence Interval
A confidence interval provides a range of values that is likely to contain the true population parameter, based on the sample data. The confidence level (e.g., 90%, 95%, or 99%) is associated with the interval and represents how confident we are that the true parameter lies within the interval.

For example, a 95% confidence interval means that if we were to take many random samples from the population and calculate a confidence interval for each sample, about 95% of those intervals would contain the true population parameter.

15.What is the margin of error, and how does it affect the confidence interval?

.The margin of error (MoE) is the amount by which we expect a sample estimate (like a sample mean or proportion) to differ from the true population parameter due to sampling variability. It quantifies the uncertainty in our estimate and is used to create the confidence interval.

In other words, the margin of error tells us how much error we might expect between our sample estimate and the true population parameter. The wider the margin of error, the less precise the estimate, and the narrower the margin of error, the more precise the estimate.

16.How is Bayes' Theorem used in statistics, and what is its significance?

.Bayes' Theorem is a fundamental concept in statistics and probability theory that describes the probability of an event occurring, based on prior knowledge of related events. It allows us to update our beliefs about a hypothesis or event in light of new evidence.

In simple terms, Bayes' Theorem helps us calculate the posterior probability of a hypothesis, given new data, by combining our prior knowledge (prior probability) with the likelihood of the observed data.

1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.































```python
import numpy as np
from scipy import stats

# Sample data (replace with your own sample values)
sample_data = [23, 21, 22, 20, 19, 24, 23, 22, 21, 20]

# Known population mean
population_mean = 22

# Population standard deviation (known or assumed)
population_std = 2.5

# Sample size
sample_size = len(sample_data)

# Sample mean
sample_mean = np.mean(sample_data)

# Standard error of the mean (SEM)
sem = population_std / np.sqrt(sample_size)

# Z-statistic calculation
z_statistic = (sample_mean - population_mean) / sem

# p-value calculation for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

# Display results
print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpretation of the result
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def type_1_type_2_error(population_mean, population_std, sample_size, true_mean, alpha=0.05):
    # Simulate sample data (under the assumption that the population mean is population_mean)
    np.random.seed(42)  # For reproducibility

    # Generate a sample under the null hypothesis (population mean)
    null_sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

    # Calculate the sample mean under the null hypothesis
    null_sample_mean = np.mean(null_sample_data)

    # Standard error of the mean (SEM)
    sem = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic under the null hypothesis
    z_stat_null = (null_sample_mean - population_mean) / sem

    # Perform a one-sample Z-test under the null hypothesis
    p_value_null = 2 * (1 - stats.norm.cdf(abs(z_stat_null)))  # Two-tailed test

    # Calculate the rejection region (critical values for two-tailed test at alpha significance level)
    critical_value = stats.norm.ppf(1 - alpha/2)  # Z-critical value for two-tailed test

    # Type I Error: Rejecting null hypothesis when it is true
    type_1_error = p_value_null < alpha

    # Now calculate Type II error by simulating data assuming the true mean is not the population mean
    # Simulate data assuming the true mean is different
    true_sample_data = np.random.normal(loc=true_mean, scale=population_std, size=sample_size)

    # Calculate the sample mean under the true alternative hypothesis
    true_sample_mean = np.mean(true_sample_data)

    # Z-statistic for the alternative hypothesis (true mean)
    z_stat_true = (true_sample_mean - population_mean) / sem

    # Calculate the p-value for this sample
    p_value_true = 2 * (1 - stats.norm.cdf(abs(z_stat_true)))  # Two-tailed test

    # Type II Error: Failing to reject null hypothesis when the true mean is different (false negative)
    type_2_error = p_value_true >= alpha

    # Print Results
    print(f"Sample Mean under Null Hypothesis: {null_sample_mean}")
    print(f"Sample Mean under True Alternative Hypothesis: {true_sample_mean}")
    print(f"Z-statistic under Null Hypothesis: {z_stat_null}")
    print(f"Z-statistic under True Mean: {z_stat_true}")
    print(f"p-value under Null Hypothesis: {p_value_null}")
    print(f"p-value under True Mean: {p_value_true}")
    print(f"Type I Error (Rejecting null when true): {type_1_error}")
    print(f"Type II Error (Failing to reject when false): {type_2_error}")

    # Plot visualization
    x = np.linspace(-4, 4, 1000)  # Range of x values for the normal distribution plot
    y = stats.norm.pdf(x, 0, 1)  # Probability density function for standard normal distribution

    plt.plot(x, y, label="Standard Normal Distribution", color='blue')

    # Shading the rejection region (Type I error region) and non-rejection region
    plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.5, label="Rejection Region (Left)")
    plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.5, label="Rejection Region (Right)")

    # Mark the calculated Z-statistics
    plt.axvline(z_stat_null, color='green', linestyle='dashed', linewidth=2, label=f"Z-statistic (Null): {z_stat_null:.2f}")
    plt.axvline(z_stat_true, color='purple', linestyle='dashed', linewidth=2, label=f"Z-statistic (True Mean): {z_stat_true:.2f}")

    # Add labels and title
    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage:
population_mean = 100   # Known population mean
population_std = 15     # Known population standard deviation
sample_size = 50        # Sample size
true_mean = 105         # True mean under the alternative hypothesis

type_1_type_2_error(population_mean, population_std, sample_size, true_mean)
```

6. Write a Python program to perform an independent T-test and interpret the results.


```python
import numpy as np
from scipy import stats

# Simulate two independent sample groups
np.random.seed(42)  # For reproducibility
group_1 = np.random.normal(loc=100, scale=15, size=30)  # Group 1: Mean=100, Std=15, Sample size=30
group_2 = np.random.normal(loc=105, scale=15, size=30)  # Group 2: Mean=105, Std=15, Sample size=30

# Perform an independent T-test
t_statistic, p_value = stats.ttest_ind(group_1, group_2)

# Display results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The two groups have significantly different means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")

```

7.Perform a paired sample T-test using Python and visualize the comparison results.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate paired sample data (e.g., pre-test and post-test scores)
np.random.seed(42)  # For reproducibility
pre_test = np.random.normal(loc=80, scale=10, size=30)  # Pre-test scores
post_test = pre_test + np.random.normal(loc=5, scale=5, size=30)  # Post-test scores (with improvement)

# Perform a paired sample T-test
t_statistic, p_value = stats.ttest_rel(pre_test, post_test)

# Display the results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the pre-test and post-test scores.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the pre-test and post-test scores.")

# Visualization: Plotting pre-test and post-test scores
plt.figure(figsize=(8, 5))

# Create a box plot for the two samples
plt.boxplot([pre_test, post_test], labels=['Pre-test', 'Post-test'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'), flierprops=dict(markerfacecolor='red', marker='o'))

# Add titles and labels
plt.title('Comparison of Pre-test and Post-test Scores')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample_data: List or numpy array of sample data.
    - confidence_level: Confidence level for the interval (default is 95%).

    Returns:
    - lower_bound: Lower bound of the confidence interval.
    - upper_bound: Upper bound of the confidence interval.
    """

    # Convert sample data to a numpy array (if it's a list)
    sample_data = np.array(sample_data)

    # Calculate sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample standard deviation
    sample_size = len(sample_data)

    # Determine the z-score for the given confidence level
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Generate random sample data

# Calculate the 95% confidence interval
lower, upper = calculate_confidence_interval(sample_data, confidence_level=0.95)

# Output the results
print(f"Sample Mean: {np.mean(sample_data)}")
print(f"95% Confidence Interval: ({lower}, {upper})")
```

11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


```python
def bayesian_inference(prior_disease, likelihood_positive_given_disease, likelihood_positive_given_no_disease):
    """
    Perform Bayesian Inference to calculate the posterior probability of having a disease given a positive test result.

    Parameters:
    - prior_disease: Prior probability of having the disease (P(Disease))
    - likelihood_positive_given_disease: Likelihood of testing positive given disease (P(Positive | Disease))
    - likelihood_positive_given_no_disease: Likelihood of testing positive given no disease (P(Positive | No Disease))

    Returns:
    - posterior_disease: Posterior probability of having the disease given a positive test result (P(Disease | Positive))
    """
    # Calculate the prior probability of not having the disease
    prior_no_disease = 1 - prior_disease

    # Calculate the marginal likelihood P(Positive)
    p_positive = (likelihood_positive_given_disease * prior_disease) + (likelihood_positive_given_no_disease * prior_no_disease)

    # Apply Bayes' Theorem to compute the posterior probability
    posterior_disease = (likelihood_positive_given_disease * prior_disease) / p_positive

    return posterior_disease

# Example values
prior_disease = 0.01  # 1% chance of having the disease (prior)
likelihood_positive_given_disease = 0.95  # 95% chance of testing positive if you have the disease
likelihood_positive_given_no_disease = 0.05  # 5% chance of testing positive if you don't have the disease

# Perform Bayesian inference
posterior_disease = bayesian_inference(prior_disease, likelihood_positive_given_disease, likelihood_positive_given_no_disease)

# Output the results
print(f"Posterior Probability of Disease given a Positive Test Result: {posterior_disease:.4f}")
```

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    - observed_data: A 2D numpy array (contingency table) representing the observed frequencies.

    Returns:
    - expected_data: A 2D numpy array representing the expected frequencies.
    """
    # Step 1: Calculate the sum of each row (marginal totals)
    row_sums = observed_data.sum(axis=1)

    # Step 2: Calculate the sum of each column (marginal totals)
    col_sums = observed_data.sum(axis=0)

    # Step 3: Calculate the total sum of all observations
    total_sum = observed_data.sum()

    # Step 4: Calculate the expected frequencies for each cell using the formula
    expected_data = np.outer(row_sums, col_sums) / total_sum

    return expected_data

# Example contingency table (observed data)
observed_data = np.array([[30, 10],  # Male: 30 prefer Product A, 10 prefer Product B
                          [20, 20]]) # Female: 20 prefer Product A, 20 prefer Product B

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the results
print("Observed Data:\n", observed_data)
print("\nExpected Frequencies:\n", expected_frequencies)
```

16. Implement an F-test using Python to compare the variances of two random samples.


```python
import numpy as np
import scipy.stats as stats

# Simulate two random samples with different variances
np.random.seed(42)
sample_1 = np.random.normal(loc=0, scale=5, size=50)  # Sample 1 with larger variance
sample_2 = np.random.normal(loc=0, scale=3, size=50)  # Sample 2 with smaller variance

# Step 1: Calculate the sample variances
var_1 = np.var(sample_1, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)
var_2 = np.var(sample_2, ddof=1)

# Step 2: Calculate the F-statistic (larger variance / smaller variance)
F_statistic = var_1 / var_2 if var_1 >= var_2 else var_2 / var_1

# Step 3: Find the p-value using the F-distribution
df1 = len(sample_1) - 1  # degrees of freedom for sample 1
df2 = len(sample_2) - 1  # degrees of freedom for sample 2
p_value = 1 - stats.f.cdf(F_statistic, df1, df2)  # For two-tailed test

# Output the results
print("Sample 1 variance:", var_1)
print("Sample 2 variance:", var_2)
print("F-statistic:", F_statistic)
print("P-value:", p_value)

# Step 4: Compare the p-value with the significance level
alpha = 0.05  # significance level

if p_value < alpha:
    print("\nWe reject the null hypothesis: The variances are significantly different.")
else:
    print("\nWe fail to reject the null hypothesis: The variances are not significantly different.")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.


```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(x1, n1, x2, n2, alpha=0.05):
    """
    Perform a Z-test for comparing two proportions.

    Parameters:
    x1, n1 : int  - Successes and total observations in Group 1
    x2, n2 : int  - Successes and total observations in Group 2
    alpha : float - Significance level (default 0.05)

    Returns:
    Z-statistic, P-value, and conclusion of hypothesis test.
    """

    # Calculate sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    p = (x1 + x2) / (n1 + n2)

    # Standard error calculation
    standard_error = np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

    # Compute Z-score
    z_statistic = (p1 - p2) / standard_error

    # Compute p-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Print results
    print(f"Group 1 Proportion: {p1:.4f} (x1={x1}, n1={n1})")
    print(f"Group 2 Proportion: {p2:.4f} (x2={x2}, n2={n2})")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in proportions.")

# Example: Two groups with different success rates
x1, n1 = 50, 200  # Group 1: 50 successes out of 200 trials
x2, n2 = 30, 180  # Group 2: 30 successes out of 180 trials

# Perform the Z-test for proportions
z_test_proportions(x1, n1, x2, n2)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variance(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two datasets.

    Parameters:
    sample1, sample2 : array-like - The two datasets
    alpha : float - Significance level (default 0.05)

    Returns:
    F-statistic, P-value, and conclusion of hypothesis test.
    """
    # Compute sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Compute F-statistic (always use larger variance as numerator)
    if var1 > var2:
        F_statistic = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        F_statistic = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    # Compute p-value
    p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

    # Print results
    print(f"Sample 1 Variance: {var1:.4f}, Sample 2 Variance: {var2:.4f}")
    print(f"F-statistic: {F_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variances.")

    # Visualization of the F-distribution and critical region
    x = np.linspace(0.1, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8,5))
    plt.plot(x, y, label="F-distribution")
    plt.axvline(F_statistic, color='r', linestyle='--', label="Observed F-statistic")

    # Shade critical region
    crit_val_right = stats.f.ppf(1 - alpha/2, df1, df2)
    crit_val_left = stats.f.ppf(alpha/2, df1, df2)

    plt.fill_between(x, y, where=(x >= crit_val_right), color='red', alpha=0.3, label="Critical Region (Right)")
    plt.fill_between(x, y, where=(x <= crit_val_left), color='red', alpha=0.3, label="Critical Region (Left)")

    plt.title("F-test: Comparing Variances")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# Generate sample data
np.random.seed(42)
sample1 = np.random.normal(50, 5, 30)  # Mean=50, Std=5, n=30
sample2 = np.random.normal(50, 8, 30)  # Mean=50, Std=8, n=30

# Perform F-test
f_test_variance(sample1, sample2)
```

27.Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulating observed data (random categorical counts)
np.random.seed(42)
observed = np.random.randint(20, 100, size=5)  # Random counts for 5 categories

# Defining expected frequencies (assumed uniform distribution)
expected = np.full_like(observed, np.mean(observed))

# Perform Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")

# Visualization
categories = range(1, len(observed) + 1)
plt.bar(categories, observed, color='blue', alpha=0.6, label="Observed")
plt.bar(categories, expected, color='red', alpha=0.6, label="Expected")
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.title("Chi-square Goodness of Fit Test")
plt.legend()
plt.show()
```
